```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='dashed')
```

```python
plt.plot(x,y,color='yellow',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='#888400',marker='*',linestyle='',markersize=15)
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend(loc='best', shadow='True', fontsize='small') #lower left

plt.grid()
```

```python
import numpy as np
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
#create an x position as it wont take strings
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit, label="Profit")

plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")

plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,  label="Profit")

plt.legend()
```

```python
plt.yticks(xpos,company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")

plt.barh(xpos-0.2,revenue, height=0.4, label="Revenue")
plt.barh(xpos+0.2,profit,  height=0.4, label="Profit")

plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], 
         rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], 
         rwidth=0.95, orientation='horizontal')
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=0.5)
plt.show()
```

```python
#Explode
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
#counterclock and angle properties

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',
        radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=False, startangle=180)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=0.5,explode=[0,0,0,0.1,0.2])
plt.show()

plt.savefig("piechart.png",bbox_inches="tight", pad_inches=2,transparent=True)
```
